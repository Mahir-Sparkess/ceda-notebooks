# Working with Xarray example data

This Notebook shows examples of reading data with Xarray, manipulating it and plotting (using Matplotlib).

Initiate matplotlib to enable plots to be shown in the Notebook.

```python
%matplotlib inline
```

Import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
```

## Read some data

Load in an example data file via Xarray's `tutorial` package...
(NOTE: This downloads small NetCDF files and caches them under: `~/.xarray_tutorial_data/`)

Read in some air temperature data, and adjust it from Kelvin to degrees C.

```python
airtemps = xr.tutorial.open_dataset('air_temperature')
airtemps
air = airtemps.air - 273.15
air.attrs = airtemps.air.attrs
air.attrs['units'] = 'deg C'
```

Take a time slice and plot the data.

```python
t = air.isel(time=slice(0, 365 * 4, 250))
g_simple = t.plot(x='lon', y='lat', col='time', col_wrap=3)
```

## Plot a line graph

Get a 1D slice through our main variable and plot as a line graph.

```python
air1d = air.isel(lat=10, lon=10)
air1d.plot()
```

## Select three lines of latitude and plot against time

This `xarray` command sub-selects and plots in one line.

```python
air.isel(lon=10, lat=[19,21,22]).plot.line(x='time')
```
